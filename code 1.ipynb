```python
def fibonacci_recursive(n):
    if n <= 1:
        return n
    else:
        return fibonacci_recursive(n - 1) + fibonacci_recursive(n - 2)

# Example usage
n_terms = 10
fibonacci_result = [fibonacci_recursive(i) for i in range(n_terms)]
print(f"Fibonacci sequence (recursive) for {n_terms} terms:")
print(fibonacci_result)

```
